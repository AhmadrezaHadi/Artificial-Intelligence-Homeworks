# Puzzle Problem
In this problem we have a board and an agent and a destination, by default the size of the board is 8 by 8. <br>
Agent must find the best path to the destination (assuming there is at least one path).<br>
Agents moves : {up, down, right, left}

First we import the required modules:

```python
import networkx as nx
from matplotlib import pyplot as plt
from classes.PuzzleProblem import PuzzleProblem
```

Then we specify wall locations, agents location and agents destination:

```python
wall_locations = [(0,6), (1,6), (2,6), (3,6), (4,6), (5,6), (6,6), (2,2), (3,2), (4,2), (5,2), (7,2), (7,3), (7,4), (6,4)]
agent_dest = (0,7)
agent_loc = (4,0)
P = PuzzleProblem(size=8, agent_loc=agent_loc, wall_loc=wall_locations, target_loc=agent_dest)
```

Now we can define our puzzle:

```python
P = PuzzleProblem(size=8, agent_loc=agent_loc, wall_loc=wall_locations, target_loc=agent_dest)
```

We can draw our puzzle with `drawPuzzle()` method:

```python
P.drawPuzzle()
```

We can remove wall nodes, since we can't go through a wall :)

```python
P.drawSearchPuzzle()
```

Finding best path using [BFS](https://en.wikipedia.org/wiki/Breadth-first_search) algorithm:

```python
P.bfs()
```

```python
P.drawPuzzle()
```

Finding best path using [Iterative Deepening Search](https://en.wikipedia.org/wiki/Iterative_deepening_depth-first_search) algorithm:

```python
P.IDS()
```

```python
P.drawPuzzle()
```
